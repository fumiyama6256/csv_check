```python
import numpy as np
import pandas as pd
import glob
import os
```

```python
#CSVが置いてある場所を指定する、番号順にソート。
csv_files = sorted(glob.glob('/Users/nagayamaikuya/Desktop/My pandas/CSV/*.csv'))
csv_files
```

```python
#ファイル名を｢file_name｣に代入してdfの1列目に挿入する。
#｢data_list｣にリスト化して代入。
data_list = []
for i in range(len(csv_files)):
    df = pd.read_csv(csv_files[i])
    file_name = os.path.basename(csv_files[i])
    df.insert(0, 'file', file_name)
    data_list.append(df)
df
```

```python
#↑で作成したcsvを結合する。indexは1から振る。
df = pd.concat(data_list).reset_index(drop=True)
df.index = np.arange(1, len(df)+1)
df
```
